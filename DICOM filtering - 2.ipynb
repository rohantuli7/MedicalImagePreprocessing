```python
#importing libraries
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import pydicom
import os
import sys
import skimage 
```

```python
#loading dicom image
from pydicom.data import get_testdata_file
ds = pydicom.dcmread('/Users/rt/Desktop/Dicom/dicom/dicom_dir/ID_0000_AGE_0060_CONTRAST_1_CT.dcm')
plt.imshow(ds.pixel_array,cmap = plt.cm.gray)
plt.show()

#storing pixel values into a list
gray = np.zeros([512,512],dtype = np.uint8)
dcm_input = ds.pixel_array
gray = cv.normalize(dcm_input, gray, 0, 255, cv.NORM_MINMAX)
gray = np.uint8(gray)

#cv.imwrite('/Users/rt/Desktop/Dicom output images/Dicom_image.jpg',gray)
```

```python
#Histogram Equalization

#Histogram before equalization
hist,bins = np.histogram(gray.flatten(),512,[0,512])
plt.hist(gray.flatten(),512,[0,512], color = 'r')
plt.xlim([0,512])
plt.show()

#histogram after equalization
gray_equalized = cv.equalizeHist(gray)
hist,bins = np.histogram(gray_equalized.flatten(),512,[0,512])
plt.hist(gray_equalized.flatten(),512,[0,512], color = 'r')
plt.xlim([0,512])
plt.show()

#image before equalization
plt.imshow(gray)
plt.show()

#image after equalization
plt.imshow(gray_equalized)
plt.show()
#cv.imwrite('/Users/rt/Desktop/Dicom output images/gray_equalized.jpg',gray_equalized)
```
